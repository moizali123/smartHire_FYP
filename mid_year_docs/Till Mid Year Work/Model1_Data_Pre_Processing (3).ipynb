# Making Data available in lists.

```python
##Make Question's List
ques_lst = []
for i in range (1,28):
  row_data = ques.iat[i,0]
  ques_lst.append(row_data)
print(ques_lst)
```

```python
##This Function will return answer's list of the given question number
def get_ans(ans,n):
  ans_lst = []
  for i in range (1,8):
    row_data = ans.iat[i,n]
    ans_lst.append(row_data)
  return ans_lst

##Making list of answers
ans_lst=[]
for j in range (0,15):
  ans_lst.append(get_ans(ans1,j))
for k in range (0,13):
  ans_lst.append(get_ans(ans2,k))

print(ans_lst)
```

# Applying NLP Steps onto Answer's List.

```python
##Applying Tokenizing onto Answers
tokenized_ans_lst_main=[]

for i in range(0,28):
  tokenized_ans_lst_inter=[]
  for j in ans_lst[i]:
    lst=word_tokenize(str(j))
    tokenized_ans_lst_inter.append(lst)
  tokenized_ans_lst_main.append(tokenized_ans_lst_inter)

print(tokenized_ans_lst_main)
  

```

```python
##Removing Stop Words from Tokenized Answers
stop_words = set(stopwords.words('english'))

without_stop_words_main = []

for i in range(0,28):
  without_stop_words_inter=[]
  for j in tokenized_ans_lst_main[i]:
    lst = []
    for k in j:
      if k not in stop_words:
        lst.append(k)
    without_stop_words_inter.append(lst)
  without_stop_words_main.append(without_stop_words_inter)

print(without_stop_words_main)
```

```python
##Applying Lemmatizing onto Answer list

def do_lemmatizing(data,pos):
  lemmatizer = WordNetLemmatizer()
  
  lemmatized_main = []
  
  for i in range(0,28):
    lemmatized_inter=[]
    for j in data[i]:
      lem = [lemmatizer.lemmatize(word , pos) for word in j]
      lemmatized_inter.append(lem)
    lemmatized_main.append(lemmatized_inter)
  return lemmatized_main

verb_lemmatized = do_lemmatizing(without_stop_words_main,'v')
adj_lemmatized = do_lemmatizing(verb_lemmatized,'a')
noun_lemmatized = do_lemmatizing(adj_lemmatized,'n')
lemmatized_ans = do_lemmatizing(noun_lemmatized,'r')

print(lemmatized_ans)
```

```python
##Applying POS Tagging
tagged_words_main = []

for i in range(0,28):
    tagged_words_inter=[]
    for j in lemmatized_ans[i]:
      tag = nltk.pos_tag(j)
      tagged_words_inter.append(tag)
    tagged_words_main.append(tagged_words_inter)

print(tagged_words_main)
```

```python
##Applying Dependency Parsing
def dependency_par(ans_lst,n):
  import spacy
  from spacy import displacy 
  nlp = spacy.load("en_core_web_sm")
  for i in range(0,n):
    for j in ans_lst[i]:
      doc = nlp(j)
      print(f"{'Node (from)-->':<15} {'Relation':^10} {'-->Node (to)':>15}\n")
      for token in doc:
        print("{:<15} {:^10} {:>15}".format(str(token.head.text), str(token.dep_), str(token.text)))

dependency_par(ans_lst,1)
```
